```python
import numpy as np
data_sq_1 = np.loadtxt('Trial_2/squat_1.csv', delimiter=",", usecols=range(3))
```

```python
x_sq_1 = np.zeros((data_sq_1.shape[0]//2))
y_sq_1 = np.zeros((data_sq_1.shape[0]//2))
z_sq_1 = np.zeros((data_sq_1.shape[0]//2))
ang_x_sq_1 = np.zeros((data_sq_1.shape[0]//2))
ang_y_sq_1 = np.zeros((data_sq_1.shape[0]//2))
ang_z_sq_1 = np.zeros((data_sq_1.shape[0]//2))
for i in range((data_sq_1.shape[0])//2):
    x_sq_1[i] = data_sq_1[(2*i),0]
    y_sq_1[i] = data_sq_1[(2*i),1]
    z_sq_1[i] = data_sq_1[(2*i),2]
    ang_x_sq_1[i] = data_sq_1[(2*i)+1,0]
    ang_y_sq_1[i] = data_sq_1[(2*i)+1,1]
    ang_z_sq_1[i] = data_sq_1[(2*i)+1,2]
```

```python
import matplotlib.pyplot as plt
t_sq_1 = np.arange(start=0, stop=x_sq_1.shape[0], step=1)
```

```python
plt.plot(t_sq_1, y_sq_1)
plt.grid()
plt.title('Squat y')
```

```python
plt.plot(t_sq_1, z_sq_1)
plt.grid(axis = 'both')
plt.title('Squat z')
```

```python
plt.plot(t_sq_1, ang_x_sq_1)
plt.grid()
plt.title('Squat Angle x')
```

```python
plt.plot(t_sq_1, ang_y_sq_1)
plt.grid()
plt.title('Squat Angle y')
```

```python
plt.plot(t_sq_1, ang_z_sq_1)
plt.grid()
plt.title('Squat Angle z')
```
