```python
import numpy as np
data_pu_1 = np.loadtxt('pushup_1.csv', delimiter=",", usecols=range(3))
```

```python
x_pu_1 = np.zeros((data_pu_1.shape[0]//2))
y_pu_1 = np.zeros((data_pu_1.shape[0]//2))
z_pu_1 = np.zeros((data_pu_1.shape[0]//2))
ang_x_pu_1 = np.zeros((data_pu_1.shape[0]//2))
ang_y_pu_1 = np.zeros((data_pu_1.shape[0]//2))
ang_z_pu_1 = np.zeros((data_pu_1.shape[0]//2))
for i in range((data_pu_1.shape[0])//2):
    x_pu_1[i] = data_pu_1[(2*i),0]
    y_pu_1[i] = data_pu_1[(2*i),1]
    z_pu_1[i] = data_pu_1[(2*i),2]
    ang_x_pu_1[i] = data_pu_1[(2*i)+1,0]
    ang_y_pu_1[i] = data_pu_1[(2*i)+1,1]
    ang_z_pu_1[i] = data_pu_1[(2*i)+1,2]
```

```python
import matplotlib.pyplot as plt
t_pu_1 = np.arange(start=0, stop=x_pu_1.shape[0], step=1)
```

```python
data_pu_2 = np.loadtxt('pushup_2.csv', delimiter=",", usecols=range(3))
```

```python
x_pu_2 = np.zeros((data_pu_2.shape[0]//2))
y_pu_2 = np.zeros((data_pu_2.shape[0]//2))
z_pu_2 = np.zeros((data_pu_2.shape[0]//2))
ang_x_pu_2 = np.zeros((data_pu_2.shape[0]//2))
ang_y_pu_2 = np.zeros((data_pu_2.shape[0]//2))
ang_z_pu_2 = np.zeros((data_pu_2.shape[0]//2))
for i in range((data_pu_2.shape[0])//2):
    x_pu_2[i] = data_pu_2[(2*i),0]
    y_pu_2[i] = data_pu_2[(2*i),1]
    z_pu_2[i] = data_pu_2[(2*i),2]
    ang_x_pu_2[i] = data_pu_2[(2*i)+1,0]
    ang_y_pu_2[i] = data_pu_2[(2*i)+1,1]
    ang_z_pu_2[i] = data_pu_2[(2*i)+1,2]
```

```python
t_pu_2 = np.arange(start=0, stop=x_pu_2.shape[0], step=1)
```

```python
data_pu_3 = np.loadtxt('pushup_3.csv', delimiter=",", usecols=range(3))
```

```python
x_pu_3 = np.zeros((data_pu_3.shape[0]//2))
y_pu_3 = np.zeros((data_pu_3.shape[0]//2))
z_pu_3 = np.zeros((data_pu_3.shape[0]//2))
ang_x_pu_3 = np.zeros((data_pu_3.shape[0]//2))
ang_y_pu_3 = np.zeros((data_pu_3.shape[0]//2))
ang_z_pu_3 = np.zeros((data_pu_3.shape[0]//2))
for i in range((data_pu_3.shape[0])//2):
    x_pu_3[i] = data_pu_3[(2*i),0]
    y_pu_3[i] = data_pu_3[(2*i),1]
    z_pu_3[i] = data_pu_3[(2*i),2]
    ang_x_pu_3[i] = data_pu_3[(2*i)+1,0]
    ang_y_pu_3[i] = data_pu_3[(2*i)+1,1]
    ang_z_pu_3[i] = data_pu_3[(2*i)+1,2]
```

```python
t_pu_3 = np.arange(start=0, stop=x_pu_3.shape[0], step=1)
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_pu_1, y_pu_1)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(t_pu_2, y_pu_2)
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(t_pu_3, y_pu_3)
plt.grid()

plt.title('y acceleration')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_pu_1, z_pu_1)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(t_pu_2, z_pu_2)
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(t_pu_3, z_pu_3)
plt.grid()

plt.title('z acceleration')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_pu_1, ang_y_pu_1)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(t_pu_2, ang_y_pu_2)
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(t_pu_3, ang_y_pu_3)
plt.grid()

plt.title('angle y')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_pu_1, ang_z_pu_1)
plt.grid(axis='both')

plt.subplot(1, 3, 2)
plt.plot(t_pu_2, ang_z_pu_2)
plt.grid(axis='both')

plt.subplot(1, 3, 3)
plt.plot(t_pu_3, ang_z_pu_3)
plt.grid(axis='both')

plt.title('angle z')
plt.show()
```
