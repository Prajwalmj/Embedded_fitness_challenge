```python
import numpy as np
data_su_1 = np.loadtxt('sit_up_1.csv', delimiter=",", usecols=range(3))
```

```python
x_su_1 = np.zeros((data_su_1.shape[0]//2))
y_su_1 = np.zeros((data_su_1.shape[0]//2))
z_su_1 = np.zeros((data_su_1.shape[0]//2))
ang_x_su_1 = np.zeros((data_su_1.shape[0]//2))
ang_y_su_1 = np.zeros((data_su_1.shape[0]//2))
ang_z_su_1 = np.zeros((data_su_1.shape[0]//2))
for i in range((data_su_1.shape[0])//2):
    x_su_1[i] = data_su_1[(2*i),0]
    y_su_1[i] = data_su_1[(2*i),1]
    z_su_1[i] = data_su_1[(2*i),2]
    ang_x_su_1[i] = data_su_1[(2*i)+1,0]
    ang_y_su_1[i] = data_su_1[(2*i)+1,1]
    ang_z_su_1[i] = data_su_1[(2*i)+1,2]
```

```python
import matplotlib.pyplot as plt
t_su_1 = np.arange(start=0, stop=x_su_1.shape[0], step=1)
```

```python
plt.plot(t_su_1, y_su_1)
plt.grid(axis = 'both')
plt.title('Sit up y')
```

```python
plt.plot(t_su_1, z_su_1)
plt.grid()
plt.title('Sit up z')
```

```python
plt.plot(t_su_1, ang_y_su_1)
plt.grid()
plt.title('Sit up Angle y')
```

```python
plt.plot(t_su_1, ang_z_su_1)
plt.grid()
plt.title('Sit up Angle z')
```
