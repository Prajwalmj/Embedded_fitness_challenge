```python
import numpy as np
data_jj_1 = np.loadtxt('jj_1.csv', delimiter=",", usecols=range(3))
```

```python
x_jj_1 = np.zeros((data_jj_1.shape[0]//2))
y_jj_1 = np.zeros((data_jj_1.shape[0]//2))
z_jj_1 = np.zeros((data_jj_1.shape[0]//2))
ang_x_jj_1 = np.zeros((data_jj_1.shape[0]//2))
ang_y_jj_1 = np.zeros((data_jj_1.shape[0]//2))
ang_z_jj_1 = np.zeros((data_jj_1.shape[0]//2))
for i in range((data_jj_1.shape[0])//2):
    x_jj_1[i] = data_jj_1[(2*i),0]
    y_jj_1[i] = data_jj_1[(2*i),1]
    z_jj_1[i] = data_jj_1[(2*i),2]
    ang_x_jj_1[i] = data_jj_1[(2*i)+1,0]
    ang_y_jj_1[i] = data_jj_1[(2*i)+1,1]
    ang_z_jj_1[i] = data_jj_1[(2*i)+1,2]
```

```python
import matplotlib.pyplot as plt
t_jj_1 = np.arange(start=0, stop=x_jj_1.shape[0], step=1)
```

```python
jj_mean_10 = np.mean(y_jj_1[:10])
jj_std_10 = np.std(y_jj_1[:10])
print(jj_mean_10)
print(jj_std_10)
jj_mean_20 = np.mean(y_jj_1[:20])
jj_std_20 = np.std(y_jj_1[:20])
print(jj_mean_20)
print(jj_std_20)
```

```python
plt.plot(t_jj_1, y_jj_1)
plt.grid()
plt.title('Jumping Jack y')
```

```python
plt.plot(t_jj_1, z_jj_1)
plt.grid()
plt.title('Jumping Jack z')
```

```python
plt.plot(t_jj_1, ang_y_jj_1)
plt.grid()
plt.title('Jumping jack Angle y')
```

```python
plt.plot(t_jj_1, ang_z_jj_1)
plt.grid()
plt.title('Jumping jack Angle z')
```
