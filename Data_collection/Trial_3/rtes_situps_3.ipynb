```python
import numpy as np
data_su_1 = np.loadtxt('Trial_3/su_1.csv', delimiter=",")
```

```python
y_su_1 = np.zeros((data_su_1.shape[0]//2))
z_su_1 = np.zeros((data_su_1.shape[0]//2))
roll_su_1 = np.zeros((data_su_1.shape[0]//2))
pitch_su_1 = np.zeros((data_su_1.shape[0]//2))
for i in range((data_su_1.shape[0])//2):
    y_su_1[i] = data_su_1[(2*i),0]
    z_su_1[i] = data_su_1[(2*i),1]
    roll_su_1[i] = data_su_1[(2*i)+1,0]
    pitch_su_1[i] = data_su_1[(2*i)+1,1]
```

```python
import matplotlib.pyplot as plt
t_su_1 = np.arange(start=0, stop=y_su_1.shape[0], step=1)
```

```python
data_su_2 = np.loadtxt('Trial_3/su_2.csv', delimiter=",")
```

```python
y_su_2 = np.zeros((data_su_2.shape[0]//2))
z_su_2 = np.zeros((data_su_2.shape[0]//2))
roll_su_2 = np.zeros((data_su_2.shape[0]//2))
pitch_su_2 = np.zeros((data_su_2.shape[0]//2))
for i in range((data_su_2.shape[0])//2):
    y_su_2[i] = data_su_2[(2*i),0]
    z_su_2[i] = data_su_2[(2*i),1]
    roll_su_2[i] = data_su_2[(2*i)+1,0]
    pitch_su_2[i] = data_su_2[(2*i)+1,1]
```

```python
t_su_2 = np.arange(start=0, stop=y_su_2.shape[0], step=1)
```

```python
data_su_3 = np.loadtxt('Trial_3/su_3.csv', delimiter=",")
```

```python
y_su_3 = np.zeros((data_su_3.shape[0]//2))
z_su_3 = np.zeros((data_su_3.shape[0]//2))
roll_su_3 = np.zeros((data_su_3.shape[0]//2))
pitch_su_3 = np.zeros((data_su_3.shape[0]//2))
for i in range((data_su_3.shape[0])//2):
    y_su_3[i] = data_su_3[(2*i),0]
    z_su_3[i] = data_su_3[(2*i),1]
    roll_su_3[i] = data_su_3[(2*i)+1,0]
    pitch_su_3[i] = data_su_3[(2*i)+1,1]
```

```python
t_su_3 = np.arange(start=0, stop=y_su_3.shape[0], step=1)
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_su_1, y_su_1)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(t_su_2, y_su_2)
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(t_su_3, y_su_3)
plt.grid()

plt.title('y acceleration')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_su_1, z_su_1)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(t_su_2, z_su_2)
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(t_su_3, z_su_3)
plt.grid()

plt.title('z acceleration')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_su_1, roll_su_1)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(t_su_2, roll_su_2)
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(t_su_3, roll_su_3)
plt.grid()

plt.title('roll angle')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_su_1, pitch_su_1)
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(t_su_2, pitch_su_2)
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(t_su_3, pitch_su_3)
plt.grid()

plt.title('pitch angle')
plt.show()
```
