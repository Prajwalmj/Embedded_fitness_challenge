```python
import numpy as np
data_sq_1 = np.loadtxt('Trial_3/sq_1.csv', delimiter=",")
```

```python
y_sq_1 = np.zeros((data_sq_1.shape[0]//2))
z_sq_1 = np.zeros((data_sq_1.shape[0]//2))
roll_sq_1 = np.zeros((data_sq_1.shape[0]//2))
pitch_sq_1 = np.zeros((data_sq_1.shape[0]//2))
for i in range((data_sq_1.shape[0])//2):
    y_sq_1[i] = data_sq_1[(2*i),0]
    z_sq_1[i] = data_sq_1[(2*i),1]
    roll_sq_1[i] = data_sq_1[(2*i)+1,0]
    pitch_sq_1[i] = data_sq_1[(2*i)+1,1]
```

```python
import matplotlib.pyplot as plt
t_sq_1 = np.arange(start=0, stop=y_sq_1.shape[0], step=1)
```

```python
data_sq_2 = np.loadtxt('Trial_3/sq_2.csv', delimiter=",")
```

```python
y_sq_2 = np.zeros((data_sq_2.shape[0]//2))
z_sq_2 = np.zeros((data_sq_2.shape[0]//2))
roll_sq_2 = np.zeros((data_sq_2.shape[0]//2))
pitch_sq_2 = np.zeros((data_sq_2.shape[0]//2))
for i in range((data_sq_2.shape[0])//2):
    y_sq_2[i] = data_sq_2[(2*i),0]
    z_sq_2[i] = data_sq_2[(2*i),1]
    roll_sq_2[i] = data_sq_2[(2*i)+1,0]
    pitch_sq_2[i] = data_sq_2[(2*i)+1,1]
```

```python
t_sq_2 = np.arange(start=0, stop=y_sq_2.shape[0], step=1)
```

```python
data_sq_3 = np.loadtxt('Trial_3/sq_3.csv', delimiter=",")
```

```python
y_sq_3 = np.zeros((data_sq_3.shape[0]//2))
z_sq_3 = np.zeros((data_sq_3.shape[0]//2))
roll_sq_3 = np.zeros((data_sq_3.shape[0]//2))
pitch_sq_3 = np.zeros((data_sq_3.shape[0]//2))
for i in range((data_sq_3.shape[0])//2):
    y_sq_3[i] = data_sq_3[(2*i),0]
    z_sq_3[i] = data_sq_3[(2*i),1]
    roll_sq_3[i] = data_sq_3[(2*i)+1,0]
    pitch_sq_3[i] = data_sq_3[(2*i)+1,1]
```

```python
t_sq_3 = np.arange(start=0, stop=y_sq_3.shape[0], step=1)
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_sq_1, y_sq_1)
plt.grid(axis = 'both')

plt.subplot(1, 3, 2)
plt.plot(t_sq_2, y_sq_2)
plt.grid(axis = 'both')

plt.subplot(1, 3, 3)
plt.plot(t_sq_3, y_sq_3)
plt.grid(axis = 'both')

plt.title('y acceleration')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_sq_1, z_sq_1)
plt.grid(axis = 'both')

plt.subplot(1, 3, 2)
plt.plot(t_sq_2, z_sq_2)
plt.grid(axis = 'both')

plt.subplot(1, 3, 3)
plt.plot(t_sq_3, z_sq_3)
plt.grid(axis = 'both')

plt.title('z acceleration')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_sq_1, roll_sq_1)
plt.grid(axis = 'both')

plt.subplot(1, 3, 2)
plt.plot(t_sq_2, roll_sq_2)
plt.grid(axis = 'both')

plt.subplot(1, 3, 3)
plt.plot(t_sq_3, roll_sq_3)
plt.grid(axis = 'both')

plt.title('roll angle')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5))


plt.subplot(1, 3, 1)
plt.plot(t_sq_1, pitch_sq_1)
plt.grid(axis = 'both')

plt.subplot(1, 3, 2)
plt.plot(t_sq_2, pitch_sq_2)
plt.grid(axis = 'both')

plt.subplot(1, 3, 3)
plt.plot(t_sq_3, pitch_sq_3)
plt.grid(axis = 'both')

plt.title('pitch angle')
plt.show()
```
