```python
import numpy as np
data_jj_1 = np.loadtxt('Trial_3/jj_1.csv', delimiter=",")
```

```python
y_jj_1 = np.zeros((data_jj_1.shape[0]//2))
z_jj_1 = np.zeros((data_jj_1.shape[0]//2))
roll_jj_1 = np.zeros((data_jj_1.shape[0]//2))
pitch_jj_1 = np.zeros((data_jj_1.shape[0]//2))
for i in range((data_jj_1.shape[0])//2):
    y_jj_1[i] = data_jj_1[(2*i),0]
    z_jj_1[i] = data_jj_1[(2*i),1]
    roll_jj_1[i] = data_jj_1[(2*i)+1,0]
    pitch_jj_1[i] = data_jj_1[(2*i)+1,1]
```
